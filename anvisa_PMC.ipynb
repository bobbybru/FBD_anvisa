```python
csv_drugs = './TA_PRECO_MEDICAMENTO.csv'
```

```python
data=pd.read_csv(csv_drugs, sep=';', index_col=False)
```

```python
df_med=data.rename(columns={'SUBSTÂNCIA':'PRINC_ATIVO'})
df_med=df_med.rename(columns={'LABORATÓRIO':'LAB'})
df_med=df_med.rename(columns={'REGIME DE PREÇO':'REG_PRECO'})
df_med=df_med.rename(columns={'APRESENTAÇÃO':'APRESENTACAO'})
df_med=df_med.rename(columns={'CLASSE TERAPÊUTICA':'CLAS_TERAPEUTICA'})
df_med=df_med.rename(columns={'TIPO DE PRODUTO (STATUS DO PRODUTO)':'TIP_PRODUTO'})
df_med=df_med.rename(columns={'RESTRIÇÃO HOSPITALAR':'RESTR_HOSP'})
df_med=df_med.rename(columns={'CÓDIGO GGREM':'CD_GGEREM'})
df_med=df_med.rename(columns={'ANÁLISE RECURSAL':'ANL_RECURSAL'})
df_med=df_med.rename(columns={'LISTA DE CONCESSÃO DE CRÉDITO TRIBUTÁRIO (PIS/COFINS)':'LS_PIS_COFINS'})
df_med=df_med.rename(columns={'COMERCIALIZAÇÃO 2019':'COMERC_2019'})
df_med=df_med.rename(columns={'EAN 1':'EAN_1'})
df_med=df_med.rename(columns={'EAN 2':'EAN_2'})
df_med=df_med.rename(columns={'EAN 3':'EAN_3'})
df_med=df_med.rename(columns={'PMC 20%':'PMC'})
df_med=df_med.rename(columns={'PF 20%':'PF'})
df_med.head(1)
```

```python
df_med.columns
```

```python
df_med_f=df_med[['CNPJ','LAB','PRINC_ATIVO','PRODUTO','APRESENTACAO','TIP_PRODUTO','TARJA','EAN_1','EAN_2','EAN_3','LS_PIS_COFINS','PF','PMC', 'REGISTRO','CD_GGEREM']]
# df_med[df_med['CNPJ']=='07.718.721/0001-80']
```

```python
df_med_f.head(3)
```

```python
#CARGA TABELA LABORATORIO
lab = df_med_f[['CNPJ','LAB']]
lab = lab.drop_duplicates()
df_lab =lab.astype(object).where(pd.notnull(lab), None)
for row in df_lab.itertuples():
#     cnpj0= ''.join(c for c in row.CNPJ if c not in '/.-')
    sql = "INSERT INTO laboratorio (cnpj, nome)  VALUES (%s, %s)"
    val = (row.CNPJ, (row.LAB))
    try:
        cursor.execute(sql, val)
    except:
        raise
cnx.commit()
sql_lab= pd.read_sql("SELECT * from laboratorio", cnx)
sql_lab.head(3)
```

```python
#CARGA TABELA REGISTRO
registro = df_med_f[['CD_GGEREM', 'REGISTRO']]
registro = registro.drop_duplicates()
df_registro =registro.astype(object).where(pd.notnull(registro), None)
df_registro = df_registro.replace(to_replace='    -     ', value=0)

for row in df_registro.itertuples():
    print(row)
    sql = "INSERT INTO registro (idRegistro, CD_GGEREM)  VALUES (%s, %s)"
    val = (row.REGISTRO ,row.CD_GGEREM)
    try:
        cursor.execute(sql, val)
    except:
        raise
cnx.commit()
sql_lab= pd.read_sql("SELECT * from registro", cnx)
sql_lab.head(3)
```

```python
sql = "DELETE FROM preco"
cursor.execute(sql)
```

```python
#CARGA TABELA MEDICAMENTO
medicam = df_med_f[['PRINC_ATIVO','PRODUTO','APRESENTACAO','TIP_PRODUTO', 'TARJA', 'CD_GGEREM']]
medicam = medicam.drop_duplicates()
df_medicam =medicam.astype(object).where(pd.notnull(medicam), None)
        
for row in df_medicam.itertuples():
    
    print(row)
    principio=str(row.PRINC_ATIVO)[0:300]
    apresentacao=str(row.APRESENTACAO)[0:150]
    sql = "INSERT INTO medicamento (principio_ativo, apresentacao, produto, tarja, tipo_produto, Registro_CD_GGEREM)  VALUES (%s, %s, %s, %s, %s, %s)"
    val = (principio ,apresentacao ,row.PRODUTO ,row.TARJA ,row.TIP_PRODUTO,  row.CD_GGEREM)
    try:
        cursor.execute(sql, val)
    except:
        raise
cnx.commit()
sql_lab= pd.read_sql("SELECT * from medicamento", cnx)
sql_lab.head(3)
```

```python
#CARGA TABELA PRECO
preco = df_med_f[['EAN_1','EAN_2','EAN_3','LS_PIS_COFINS', 'PF', 'PMC', 'PRINC_ATIVO', 'APRESENTACAO', 'CD_GGEREM', 'PRODUTO']]
preco = preco.drop_duplicates()
df_preco =preco.astype(object).where(pd.notnull(preco), None)
df_preco = df_preco.replace(to_replace='    -     ', value=0)
# df_preco      

for row in df_preco.itertuples():
    
    print(row)
    if row.PF is None:
        pf0=0
    else:
        pf0=(row.PF).replace(',', '.')
    pf=float(pf0)
    if row.PMC is None:
        pmc0=0
    else:
        pmc0=(row.PMC).replace(',', '.')
    pmc=float(pmc0)
    principio=str(row.PRINC_ATIVO)[0:300]
    apresentacao=str(row.APRESENTACAO)[0:150]
    sql = "INSERT INTO preco (ean_1, ean_2, ean_3, ls_pis_cofins, pf, pmc, Medicamento_principio_ativo, Medicamento_apresentacao, Medicamento_Registro_CD_GGEREM, Medicamento_produto)  VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
    val = (row.EAN_1, row.EAN_2, row.EAN_3, row.LS_PIS_COFINS, pf, pmc,  principio ,apresentacao ,row.CD_GGEREM, row.PRODUTO)
    try:
        cursor.execute(sql, val)
    except:
        raise
cnx.commit()
sql_lab= pd.read_sql("SELECT * from preco", cnx)
sql_lab.head(3)
```

```python
#CARGA TABELA RELACAO LABORATORIO X MEDICAMENTO
relac = df_med_f[['CNPJ','PRINC_ATIVO','APRESENTACAO','CD_GGEREM', 'PRODUTO']]
relac = relac.drop_duplicates()
df_relac =relac.astype(object).where(pd.notnull(relac), None)
df_relac = df_relac.replace(to_replace='    -     ', value=0)

for row in df_relac.itertuples():
    
    print(row)
    principio=str(row.PRINC_ATIVO)[0:300]
    apresentacao=str(row.APRESENTACAO)[0:150]
    principio=str(row.PRINC_ATIVO)[0:300]
    apresentacao=str(row.APRESENTACAO)[0:150]
    sql = "INSERT INTO laboratorio_has_medicamento (Laboratorio_cnpj, Medicamento_principio_ativo, Medicamento_apresentacao, Medicamento_Registro_CD_GGEREM, Medicamento_produto)  VALUES (%s, %s, %s, %s, %s)"
    val = (row.CNPJ, principio, apresentacao, row.CD_GGEREM, row.PRODUTO)
    try:
        cursor.execute(sql, val)
    except:
        raise
cnx.commit()
sql_lab= pd.read_sql("SELECT * from laboratorio_has_medicamento", cnx)
sql_lab.head(3)
```
